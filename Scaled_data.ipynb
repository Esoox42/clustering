```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.cluster import SpectralClustering
from sklearn.cluster import KMeans
from sklearn.preprocessing import normalize
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
tsne = TSNE(n_components=2, perplexity=30)
```

```python
df = pd.read_csv('test.csv')
df = df.drop(columns=['user_id'])
df.head()
```

```python
df.hist()
plt.show()
```

```python
df_tsne = tsne.fit_transform(df)
```

```python
scaler = MinMaxScaler()
```

```python
scaled_data = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled_data, columns=['col1_norm', 'col2_norm', 'col3_norm'])
scaled_df = scaled_df.rename(columns={'col1_norm': 'recency', 'col2_norm': 'monetary', 'col3_norm': 'frequency'})
scaled_df.head()
```

```python
scaled_df_tsne = tsne.fit_transform(scaled_df)
```

```python
std_scaler = StandardScaler()
```

```python
std_scaled_data = std_scaler.fit_transform(df)
std_scaled_df = pd.DataFrame(std_scaled_data, columns=['col1_norm', 'col2_norm', 'col3_norm'])
std_scaled_df = std_scaled_df.rename(columns={'col1_norm': 'recency', 'col2_norm': 'monetary', 'col3_norm': 'frequency'})
std_scaled_df.head()
```

```python
std_scaled_df_tsne = tsne.fit_transform(std_scaled_df)
```

```python
plt.figure(figsize=(4,4))
plt.scatter(df_tsne[:, 0], df_tsne[:, 1])
plt.title('Dataframe')
plt.show()

plt.figure(figsize=(4,4))
plt.scatter(scaled_df_tsne[:, 0], scaled_df_tsne[:, 1])
plt.title('Scaled dataframe')
plt.show()

plt.figure(figsize=(4,4))
plt.scatter(std_scaled_df_tsne[:, 0], std_scaled_df_tsne[:, 1])
plt.title('Standardised scaled dataframe')
plt.show()
```
